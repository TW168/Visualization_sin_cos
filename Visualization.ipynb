## Visualization

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
plt.style.use('classic')
%matplotlib inline
!python --version
```

Plot sin and cos in a figure

```python
# get 256 evenly spaced numbers from -3.1415 to 3.1415 
x= np.linspace(-np.pi, np.pi, 256, endpoint=True)
C, S= np.cos(x), np.sin(x)

```

```python
# create firgure object
fig= plt.figure()

# create canvas
ax= fig.add_axes([0,0,1,1])
# make the plot 
ax.plot(x,C, label='Cos(x)')
ax.plot(x,S, label='Sin(x)')

# choose best location for legend 
ax.legend(loc=0)
ax.set_title('Cos(x) and Sin(x)')
```
